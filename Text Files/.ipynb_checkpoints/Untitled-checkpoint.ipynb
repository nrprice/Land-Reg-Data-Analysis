```python
def plotly_compare_scatter(dataframe, n_group, x_name, y_name, z_names, time_interval='D', rolling=['N', 0]):
    
    # Creates plotly graph object
    figure = go.Figure()
    # Sorts the dataframe to ensure data is in chronological order
    dataframe = dataframe.sort_values(x_name)
    # Finds the first and last date present in the data
    first_date = dataframe[x_name].min()
    last_date = dataframe[x_name].max()

    # Z names is the category to compare. I.e property_type or county_name
    for z_name in list (dataframe[z_names].unique()):
        # Creates a new dataframe for each loop to avoid appending from the previous loop
        new_dataframe = dataframe.loc[dataframe[z_names] == z_name]
        final_dataframe = pd.DataFrame(columns=[x_name, y_name, z_names])
        
        # Checks to see if we want the rolling average calculated. 
        if rolling == 'Y':
            
            # Pandas date range from the first_date and last_date in the data    
            # Frequency is specified using time_interval argument - D is days, Y is years, M is months             
            date_range = pd.date_range(first_date, last_date, freq=time_interval)
        
            # Allows us to pass the correct increment to the for loop             
            date_step_dict = {"D": 7,
                              "M": 1,
                              "Y": 1}
            
            for i in range(0, len(date_range) - date_step_dict[time_interval], date_step_dict[time_interval]):
                
                # If time_interval is 'D'                 
                # date_range[i] and date_range[i date_step_dict[time_interval]] produce dates seven days apart:
                start = date_range [i] 
                end = date_range[i + date_step_dict[time_interval]]

                # creates a new slice for every time period            
                df_slice = new_dataframe.loc[(new_dataframe[x_name] >= start) & (new_dataframe[x_name] <= end)]
                
                # Takes the median of the y values and the earliest date in the slice.                 
                y_val = df_slice[y_name].median()
                x_val = df_slice[x_name].max()
                
                # Creates a new dataframe with appropriate column names
                vals = pd.DataFrame([[x_val, y_val, z_name]], columns=list(final_dataframe.columns))

                # Appends this slice to the rest of the slices for this category
                final_dataframe = final_dataframe.append(vals)
                
                
        # If rolling average is not selected (and for how long it took to implement, it should be selected everytime...)                 
        else:
            
            # Similar logic to above, except uses the value from n_group to slice n number of rows. 
            for i in range(n_group, len(new_dataframe), n_group):

                # Creates a new slice of rows of n_group length                 
                df_slice = new_dataframe.iloc[i - n_group: i]

                # Takes the median of the y values and the earliest date in the slice.                 
                y_val = df_slice[y_name].median()
                x_val = df_slice[x_name].min()
                
                # Creates a new dataframe with appropriate column names
                vals = pd.DataFrame([[x_val, y_val, z_name]], columns=list(final_dataframe.columns))
                # Appends this slice to the rest of the slices
                final_dataframe = final_dataframe.append(vals)
        
        figure.add_trace(go.Scatter(x=final_dataframe[x_name], y=final_dataframe[y_name], mode='lines+markers', name=property_type_dict[z_name]))
        
    # Updates the plotly graph object to have a specified number of xticks     
    figure.update_xaxes(nticks=20)
    return figure
```

```python
def create_process_control_chart(dataframe, n_group, x_name, y_name, title='', time_interval='Y', rolling=['N'], y_lim='N'):
    
    if rolling[0] == 'Y':
        dataframe['rolling_average'] = dataframe['price'].rolling(rolling[1]).mean()
        y_name = 'rolling_average'
    
    figure = plt.subplot()

    dataframe = dataframe.sort_values(x_name)

    first_date = dataframe[x_name].min()
    last_date = dataframe[x_name].max()

    print('min and max dates from data')
    print(first_date, last_date)

    final_dataframe = pd.DataFrame(columns=[x_name, y_name])

    for i in range(n_group, len(dataframe), n_group):
        df_slice = dataframe.iloc[i - n_group: i]

        y_val = df_slice[y_name].median()

        x_val = df_slice[x_name].min()

        vals = pd.DataFrame([[x_val, y_val]], columns=list(final_dataframe.columns))

        final_dataframe = final_dataframe.append(vals)

    date_range = pd.date_range(start=first_date,
                               end=last_date + relativedelta(years=1),
                               freq=time_interval, closed=None)

    print('min and max ranges from pd.date_range')
    print(date_range[0], date_range[1])

    figure.plot(final_dataframe[x_name],
                final_dataframe[y_name],
                marker='o', linestyle='-', alpha=0.3)

    plt.xlim((first_date, last_date))
    

    for n in range(1, len(list(date_range)), 1):
        try:
            # Central Line
            mean = dataframe[y_name].loc[(dataframe[x_name] > date_range[n - 1]) &
                                         (dataframe[x_name] < date_range[n])].mean()

            figure.hlines(y=mean, xmin=date_range[n - 1], xmax=date_range[n], color='r', linestyle='-')

            # Standard Deviation
            sample_std = dataframe[y_name].loc[(dataframe[x_name] > date_range[n - 1]) &
                                               (dataframe[x_name] < date_range[n])].std()

            # Upper control line
            UCL = mean + sample_std * 3
            figure.hlines(y=UCL, xmin=date_range[n - 1], xmax=date_range[n], linestyle='--')

            # Lower control line
            LCL = mean - sample_std * 3
            figure.hlines(y=LCL, xmin=date_range[n - 1], xmax=date_range[n], linestyle='--', color='black')
        
        except IndexError:
            continue
    if y_lim == 'Y':
        plt.ylim(bottom=0)
    
    x_ticks = [x.year for x in date_range]
    plt.xticks(date_range, x_ticks, rotation=0)
    
    plt.suptitle(title.title(), y=0.95)
    plt.ylabel(y_name.replace("_", " ").title())
    plt.xlabel(x_name.replace("_", " ").title())

    return figure
```

```python
def median_change_across_time(dataset, area, target, subtarget=None, n=5):
    global year_to_check
    

    if subtarget == None:
        pass

    if n < 0:
        names = dataset.groupby([area, 'year_sold'])[target].median().reset_index().groupby(area)[target].sum().sort_values(ascending=True)

    if n > 0:
        names = dataset.groupby([area, 'year_sold'])[target].median().reset_index().groupby(area)[target].sum().sort_values(ascending=False)
    
    names = list (names.head(abs(n)).index)

    df = dataset.groupby([area, 'year_sold'])[target].median().reset_index()

    figure = plt.subplots()
    for name in names:
        
        vals = df[df[area] == name][target]
        years = df[df[area] == name]['year_sold']
        plt.plot(years, vals)

    plt.legend(names)
    plt.title( f'Median {target} from 1995 to {year_to_check} for {area}'.title().replace("_", ' '))
    plt.xticks(range(1995, year_to_check, 2))
    return figure
    
    

```

```python
def price_predictor(property_type_argument, location_argument, area_argument, date_to_sell_argument,
                    dataset=full_merged_df):
    
    
    scaler = MinMaxScaler()
    
    date_to_sell_argument = dt.strptime(date_to_sell_argument, "%Y-%m-%d")

    if len(dataset.loc[dataset[area_argument] == location_argument]) == 0:
        print(f"{location_argument} not found within {area_argument}")

    else:

        # Creates dataframe to be used during function, and variables for the relevent columns 
        dataframe = dataset.loc[(dataset[area_argument] == location_argument)].copy()

        price = dataframe['price']
        date_sold = dataframe['date_sold']
        date_sold_num = dataframe['date_sold'].map(dt.toordinal)
        scaler.fit(date_sold_num.values.reshape(1, -1))
        date_sold_num = date_sold_num.apply(lambda x: scaler.transform(x))
        property_type = dataframe['property_type']

        area = dataframe[area_argument]
        
        for prop in dataframe['property_type'].unique().copy():
            dataframe[prop] = np.where(dataframe['property_type'] == prop, 1, 0).copy()
        for age in dataframe['old/new'].unique():
            dataframe['new'] = np.where(dataframe['old/new'] == 'N', 1, 0)

        
        house_features_dict = {'T': 'house_terrace', 'S': 'house_semi_detached', 'F': 'flat', 'D': 'house_detached'}
        
        dataframe = dataframe.rename(columns=house_features_dict)
       
        terrace_indicator = dataframe['house_terrace']
        semi_indicator = dataframe['house_semi_detached']
        flat_indicator = dataframe['flat']
        detached_indicator = dataframe['house_detached']
        old_new_indicator = dataframe['new']
    
        

        features_list = list(zip(date_sold_num, terrace_indicator, semi_indicator, flat_indicator, detached_indicator, old_new_indicator))
        
        X_train, X_test, y_train, y_test = train_test_split(features_list, price, test_size=0.20)

        best_k = 1
        best_score = 0

        for k in range(1, len(X_train), 4):

            model = KNeighborsRegressor(n_neighbors=k)

            model.fit(X_train, y_train)

            score = model.score(X_test, y_test)

            plt.scatter(k, score)
            plt.ylabel('Score')
            plt.xlabel('N Neighbors')

            if k % 9 == 0:
                print (k, score)

            if score < best_score and k > 201:
                break

            if score > best_score:
                best_score = score
                best_k = k

            plt.title(f"N_Neighbors vs Accuracy Score")
            plt.ylim(0, 1)

        model = KNeighborsRegressor(n_neighbors=best_k)

        model.fit(X_train, y_train)

        score = model.score(X_test, y_test)
        print(f"Best K value is {best_k} and the best score is {score}")

        year_val = date_to_sell_argument.toordinal()


#       Some way of saying the prop type
        
        predicted_price = int(model.predict([[year_val]]))
    
        predicted_year = date_to_sell_argument.year

        actual = int(dataframe.loc[(dataframe['property_type'] == property_type_argument) &
                                   (dataframe['year_sold'] == predicted_year)].groupby('property_type')['price'].mean())

        difference = actual - predicted_price
        dataframe.loc[dataframe[area_argument] == location_argument].plot(kind='scatter', x='date_sold', y='price', by='property_type', title=f'Price vs Date Sold for {location_argument}'.title())
        plt.scatter(date_to_sell_argument, predicted_price, color='r')
        plt.show();

        print(f'Predicted price for {predicted_year}: £{predicted_price}')
        print(f'Mean for {predicted_year}: £{(actual)}')
        print(f'Difference of £{(predicted_price - actual)} or {int((difference / actual) * 100)}%')


```

```python
def total_pct_change(group, start_year=start_year, end_year=end_year):
    v1 = float (group.loc[group['year_sold'] == start_year]['price'])
    v2 = float (group.loc[group['year_sold'] == end_year]['price'])
    
    total_pct_change = ((v2 - v1) / v2) * 100
    
    return total_pct_change
```

```python
# Function to find the mean price for a specific percentile, as defined by the percentile_dataset

def find_mean_price(dataset, percentile_df):
    df = pd.DataFrame()
    for i in range(0, (end_year - start_year) + 1, 1):
        temp = dataset.loc[(dataset['price'] <= percentile_df['price'][i]) & (dataset['year_sold'] == percentile_df['year_sold'][i])]
        df = df.append(temp)
    return df.groupby(['year_sold'])['price'].mean().reset_index()
```
